Task 3: Use column transformers to transform the appropriate columns in the appropriate ways.

3(a): Select the categorical columns and the numerical columns.

```python
from sklearn.compose import make_column_selector
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
# Select the categorical columns.
cat_selector = make_column_selector(dtype_include = "object")
cat_selector(X_train)
```

```python
# Select numerical columns
num_selector = make_column_selector(dtype_include = "float")
num_selector(X_train)
```

```python
train_cat_data = X_train[cat_selector(X_train)]
test_cat_data = X_test[cat_selector(X_test)]
train_cat_data
```

3(b): Use a OneHotEncoder to encode the categorical columns.

```python
# We have only one categorical column (Sex) which we will encode.
# Instantiate one hot encoder.
# Sparse and handle_unknown parameters are set to change the default behaviour of the encorder
ohe_encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
#fit the OneHotEncoder on the training data
ohe_encoder.fit(train_cat_data)
#transform both the training and the testing data
train_ohe = ohe_encoder.transform(train_cat_data)
test_ohe = ohe_encoder.transform(test_cat_data)
```

```python
train_ohe
```

```python
test_ohe
```

```python
#convert Numpy arrays to dataframe, extract new column names from encoder
#set prefixes to original column names
ohe_column_names = ohe_encoder.get_feature_names_out(train_cat_data.columns)
# The above method will get geature names from the onehotencoded data

# Create a dataframe
train_ohe = pd.DataFrame(train_ohe, columns=ohe_column_names)
test_ohe = pd.DataFrame(test_ohe, columns=ohe_column_names)
train_ohe
```

Concatenate the One-hot Encoded Categorical Features with the Numerical ones.

```python
# create a numeric selector
num_selector = make_column_selector(dtype_include='number')
```

```python
# Slice the numeric columns from the train and test set
train_nums = X_train[num_selector(X_train)].reset_index(drop=True)
test_nums = X_test[num_selector(X_test)].reset_index(drop=True)
```

```python
# Combine the encoded categorical and numerical dataframes
X_train_processed = pd.concat([train_nums, train_ohe], axis=1)
X_test_processed = pd.concat([test_nums, test_ohe], axis=1)
X_train_processed.head(1996)
```
